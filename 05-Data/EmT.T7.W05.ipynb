```python
import pandas as pd
import numpy as np
import seaborn as sns
import random 

from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
ldw = pd.read_csv("london_weather.csv")
ldw.shape
```

```python
ldw = ldw.dropna()
ldw.shape
```

```python
sns.heatmap(ldw.corr())
```

```python
ldw.plot.scatter(x=["sunshine"],y=["global_radiation"])
```

```python
sns.pairplot(ldw)
```

## Tạo Model LinearRegression

```python
lnModel = linear_model.LinearRegression()
```

```python
# train

lnModel.fit( ldw[ [ "sunshine" ]  ],  ldw[ "global_radiation" ])
print( lnModel.coef_ , lnModel.intercept_ )
```

```python
ldw.plot.scatter(x=["sunshine"],y=["global_radiation"])

x = np.array([0,16])
y = (lnModel.coef_[0]) * x + (lnModel.intercept_)

plt.plot(x, y, color="red")
```
